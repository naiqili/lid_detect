```python
import numpy as np  
from sklearn import datasets
NB_SAMPLES = 500


def toy_circle(centers, noise=.05):
    nb_centers = centers.shape[0]
    X_r = np.array([]).reshape(-1, 2)
    X_g = np.array([]).reshape(-1, 2)
    y_r = np.array([]).reshape(-1, 1)
    y_g = np.array([]).reshape(-1, 1)
    for idx, center in enumerate(centers):
        X, y = datasets.make_circles(n_samples=NB_SAMPLES, factor=.6, noise=noise)
        X = 0.5 * X + center
        y = np.where(y == 0, idx, idx + nb_centers)
        X_r = np.vstack((X_r, X[y == idx]))
        y_r = np.vstack((y_r, y[y == idx].reshape(-1, 1)))
        X_g = np.vstack((X_g, X[y == idx + nb_centers]))
        y_g = np.vstack((y_g, y[y == idx + nb_centers].reshape(-1, 1)))
    return X_r, y_r[:, 0], X_g, y_g[:, 0] - nb_centers
```

```python
centers = np.array([[0, 0]])
X_circle, _, _, _ = toy_circle(centers)
```
